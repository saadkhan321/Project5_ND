```python
import pickle

enron_data = pickle.load(open("../final_project/final_project_dataset.pkl", "r"))

print 'People in the dataset:', len(enron_data)

print '-------------------------------------------------------------------------------------------------------------'

print 'Features available per person:', sum(len(v) for v in enron_data.itervalues())/len(enron_data)

## Alaternate approach

for v in enron_data.itervalues():
    c = len(v)
print 'Features available per person:', c

print '-------------------------------------------------------------------------------------------------------------'

count = 0
for k, v in enron_data.iteritems():
    if(enron_data[k]["poi"]==1):
        count +=1
    else:
        continue
print 'POIs that are of interest in the E+F dataset:', count

print '-------------------------------------------------------------------------------------------------------------'

with open('poi_names.txt') as enron_names:
    a = -2
    for line in enron_names:
        a +=1
print 'POIs names in the POI names txt file:', a

print '-------------------------------------------------------------------------------------------------------------'

print 'Total value of the stock belonging to James Prentice:',enron_data["PRENTICE JAMES"]["total_stock_value"]

print '-------------------------------------------------------------------------------------------------------------'

print 'Email messages from Wesley Colwell to persons of interest:',enron_data["COLWELL WESLEY"]["from_this_person_to_poi"]

print '-------------------------------------------------------------------------------------------------------------'

print 'Stock options exercised by Jeffrey Skilling:',enron_data["SKILLING JEFFREY K"]["exercised_stock_options"]

print '-------------------------------------------------------------------------------------------------------------'

### Largest take home amount

js = enron_data["SKILLING JEFFREY K"]["total_payments"]
kl = enron_data["LAY KENNETH L"]["total_payments"]
af = enron_data["FASTOW ANDREW S"]["total_payments"]

if (js>kl and js>af):
    print'Largest total payment was taken by Jeff Skilling:', js
elif (af>js and af>kl): 
    print 'Largest total payment was taken by Andy Fastow:', af
elif (kl>js and kl>af):    
    print 'Largest total payment was taken by Ken Lay:', kl

print '-------------------------------------------------------------------------------------------------------------'

salaried_people = 0
for k,v in enron_data.iteritems():
    if (enron_data[k]["salary"]!= 'NaN'):
        salaried_people += 1
    else:
        continue
print 'Number fo people salaried by ENRON:', salaried_people

print '-------------------------------------------------------------------------------------------------------------'

email_ids = 0
for k,v in enron_data.iteritems():
    if (enron_data[k]["email_address"]!= 'NaN'):
        email_ids += 1
    else:
        continue
print 'Number fo people with valid email addresses:', email_ids

print '-------------------------------------------------------------------------------------------------------------'

no_total_payments = 0
for k,v in enron_data.iteritems():
    if (enron_data[k]["total_payments"] == 'NaN'):
        no_total_payments += 1
    else:
        continue
print 'Number fo people with no payment information:', no_total_payments

print '-------------------------------------------------------------------------------------------------------------'

print 'Percentage of people wih no payment information:', (float(no_total_payments)/float(len(enron_data)))*100

print '-------------------------------------------------------------------------------------------------------------'

## How many people in the E+F dataset (as it currently exists) have “NaN” for their total payments? What percentage of 
## people in the dataset as a whole is this?

## How many POIs in the E+F dataset have “NaN” for their total payments? What percentage of POI’s as a whole is this?

print '-------------------------------------------------------------------------------------------------------------'

n = 0
people_dataset = []
for k in enron_data.keys():
    if (enron_data[k]["total_payments"] == 'NaN'):
        ##print k
        people_dataset.append(k)
        ##people_dataset[n] = k
        people = people_dataset[n].split(" ")
        people_dataset[n] = people[1] + " " + people[0]
        people_dataset[n] = "".join(people_dataset[n]).lower()
        n += 1
    else:
        continue

##print people_dataset


POIs_dataset = []
with open('poi_names.txt') as enron_names:
    a = 0
    for line in enron_names:
        if line.startswith('(n)') or line.startswith('(y)') :
            ##print line
            POIs_dataset.append(line[4:])
            POIs = POIs_dataset[a].split(",")
            
            ##print POIs
            POIs_dataset[a] = POIs[1][1:].replace("\n","") + " " + POIs[0]
            POIs_dataset[a] = "".join(POIs_dataset[a]).lower()
            a += 1
        else:
            continue
       
##print POIs_dataset

print 'Number of POIs wih no payment information:',len(' '.join(list(set(people_dataset) & set(POIs_dataset))))


print '-------------------------------------------------------------------------------------------------------------'

stock_value = 0
for k,v in enron_data.iteritems():
    if (enron_data[k]["total_stock_value"] == 'NaN') and (enron_data[k]["poi"]==1):
        stock_value += 1
    else:
        continue
print 'Number fo people with no stock value:', stock_value


```
